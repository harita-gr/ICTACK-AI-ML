5. Perform meaningful visualizations using the dataset.Bring at least 3 visualizations

1. Histogram showing Age

This visualization can help you observe any pattern in the distribution of ages in the dataset

```python
plt.hist(data['Age'],color="grey",rwidth=0.95)
plt.title("Age")
plt.show()
```

2. Scatterplot of Glucose vs BMI

This visualization can help you observe any patterns or relationships between 'Glucose' and 'BMI' and see if there's a distinguishable pattern based on the outcome.

```python
sns.scatterplot(x='Glucose', y='BMI', hue='Outcome', data=data)
plt.title('Glucose vs BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
```

3. Boxplot of Age by Outcome

This visualization helps you compare the distribution of ages between diabetic and non-diabetic individuals, and the presence of any outliers can be easily identified.

```python
sns.boxplot(x='Outcome', y='Age', data=data)
plt.title('Age by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Age')
plt.show()
```
